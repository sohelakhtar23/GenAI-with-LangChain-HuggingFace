```python
import nltk
nltk.download('gutenberg')
from nltk.corpus import gutenberg

data = gutenberg.raw('shakespeare-hamlet.txt')

with open('hamlet.txt', 'w') as f:
    f.write(data)
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

tokenizer = Tokenizer()

with open('hamlet.txt', 'r') as f:
    text = f.read().lower()
tokenizer.fit_on_texts([text])

```

```python
print(tokenizer.word_index)
```

```python
text
```

```python
ip_sequences = []
for line in text.split('\n'):
    text_tokens = tokenizer.texts_to_sequences([line])[0]
    for i in range(1, len(text_tokens)):
        n_gram_sequence = text_tokens[:i + 1]
        ip_sequences.append(n_gram_sequence)
max_sequence_length = max(len(x) for x in ip_sequences)
```

```python
ip_sequences
```

```python
import numpy as np
padded_sequences = np.array(pad_sequences(ip_sequences, maxlen=max_sequence_length, padding='pre'))
padded_sequences
```

```python
total_words = len(tokenizer.word_index) + 1
print(f'Total words: {total_words}')
```

```python
x, y = padded_sequences[:, :-1], padded_sequences[:, -1]
print(x)
print(type(x))
```

```python
print(y)
```

```python
import tensorflow as tf
y = tf.keras.utils.to_categorical(y, num_classes=total_words)
print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
print(f'Train shape: {x_train.shape}, {y_train.shape}')
```

```python
## Train our LSTM RNN

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,LSTM,Dense,Dropout,GRU

## Define the model
model=Sequential()
model.add(Embedding(total_words,100,input_length=max_sequence_length-1))
model.add(LSTM(150,return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(100))
model.add(Dense(total_words,activation="softmax"))

# #Compile the model
model.compile(loss="categorical_crossentropy",optimizer='adam',metrics=['accuracy'])
model.summary()
```

```python
## Train the model
history=model.fit(x_train,y_train,epochs=50,validation_data=(x_test,y_test),verbose=1)
```

```python
model.summary()
```

```python
## Save the model
model.save("next_word_lstm.h5")
## Save the tokenizer
import pickle
with open('tokenizer.pickle','wb') as handle:
    pickle.dump(tokenizer,handle,protocol=pickle.HIGHEST_PROTOCOL)
```
